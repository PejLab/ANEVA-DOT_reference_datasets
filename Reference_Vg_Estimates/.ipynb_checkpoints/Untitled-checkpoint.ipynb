```python
#make a set of all genes in V8
vg_genes = []

for file in files:
        
        f = pd.read_csv('Vg_GTEx_v8/' + file, sep='\t')
        vg_genes = vg_genes + list(f.GeneID.values)
        
vg_genes = set(vg_genes)
```

```python
vgs = pd.DataFrame({'GENEID' : list(vg_genes)})

isf = 0
for file in files:
        
        f = pd.read_csv('Vg_GTEx_v8/' + file, sep='\t')
      
        f.columns = ['genename',file, 'dgs']
        f = f.drop(columns=['dgs'])
        vgs = pd.merge(vgs,f, left_on=['GeneID'], right_on=['genename'] ,how='outer')
        vgs = vgs.drop(columns=['genename'])
    
```
